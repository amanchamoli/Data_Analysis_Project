```python
def compare_dishes(df, dish1, dish2):
    d1 = df.loc[df['name'] == dish1, 'total_time'].iloc[0]
    d2 = df.loc[df['name'] == dish2, 'total_time'].iloc[0]

    print(f"{dish1}: {d1} minutes")
    print(f"{dish2}: {d2} minutes")

    if d1 < d2:
        print(f"\n {dish1} is faster to cook")
    elif d1 > d2:
        print(f"\n {dish2} is faster to cook")
    else:
        print("\n Both take the same time")


compare_dishes(df, "Amti", "Alu Pitika")
```
