```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns
```

```python
df = pd.read_csv(r"D:\CampusX\Project-World\Ecommerce\Data\Ecommerce_cleaned.csv")
```

```python
df
```

UNIVARIAN ANALYSIS


```python
df["customer_id"].value_counts()
```

```python
df['customer_state'].value_counts().plot(
    kind='bar',
    figsize=(10,5)
)

plt.xlabel("Customer State")
plt.ylabel("Number of Customers")
plt.title("Customer Distribution by State")
plt.xticks(rotation=45)
plt.show()
```

```python
df['payment_type'].value_counts().plot(
    kind='bar',
    figsize=(7,4)
)

plt.xlabel("Payment Type")
plt.ylabel("Number of Orders")
plt.title("Distribution of Payment Types")
plt.xticks(rotation=45)
plt.show()
```

```python
df['rating'].value_counts().sort_index().plot(
    kind='pie',
    figsize=(6,6),
    autopct='%1.1f%%',
    startangle=90
)

plt.ylabel("")  # extra y-label remove
plt.title("Rating Distribution")
plt.show()
```

```python
df['order_status'].value_counts().plot(
    kind='bar',
    figsize=(7,4)
)

plt.xlabel("Order Status")
plt.ylabel("Number of Orders")
plt.title("Distribution of Order Status")
plt.xticks(rotation=0)
plt.show()
```

```python
pd.cut(
    df['price'],
    bins=[0, 200, 500, 1000, 10000],
    labels=['<200', '200-500', '500-1000', '1000+']
).value_counts().sort_index().plot(
    kind='bar',
    figsize=(7,4)
)

plt.xlabel("Price Range")
plt.ylabel("Number of Products")
plt.title("Product Price Distribution")
plt.show()
```

```python
df['category'].value_counts().plot(
    kind='bar',
    figsize=(10,5)
)

plt.xlabel("Product Category")
plt.ylabel("Number of Products")
plt.title("Top 10 Product Categories by Count")
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python

df['seller_state'].value_counts().plot(
    kind='bar',
    figsize=(10,5)
)

plt.xlabel("Seller State")
plt.ylabel("Number of Sellers")
plt.title("Number of Sellers by State")
plt.xticks(rotation=45)
plt.show()
```

```python
pd.cut(
    df['total_payment'],
    bins=[0, 200, 500, 1000, 10000],
    labels=['<200', '200-500', '500-1000', '1000+']
).value_counts().plot(
    kind='bar',
    figsize=(7,4)
)

plt.xlabel("Total Payment Range")
plt.ylabel("Number of Orders")
plt.title("Distribution of Total Payment Amount")
plt.show()
```

```python
pd.cut(
    df['delivery_charge'],
    bins=[0, 25, 50, 100, 200, 300],
    labels=['<25', '25-50', '50-100', '100-200', '200-300']
).value_counts().plot(
    kind='bar',
    figsize=(8,4)
)

plt.xlabel("Delivery Charge Range")
plt.ylabel("Number of Orders")
plt.title("Distribution of Delivery Charges")
plt.show()
```

```python
df.groupby("customer_state")["price"].sum().sort_values(ascending=False).plot(kind='bar', figsize=(8,5))

plt.xlabel("Customer_State")
plt.ylabel("Price")
plt.title("Total Price per Customer State ")
plt.show()

```

```python
df.groupby("seller_state")["price"].sum().sort_values(ascending=False).plot(kind='bar', figsize=(8,5))

plt.xlabel("Seller State")
plt.ylabel("Price")
plt.title("Total Price per Seller State")

plt.gca().yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))

plt.show()
```

```python
df.groupby("customer_state")["price"].sum().sort_values(ascending=False).plot(kind='bar', figsize=(8,5))

plt.xlabel("Customer_State")
plt.ylabel("Price")
plt.title("Total Price per Customer State")

plt.gca().yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))

plt.show()
```

```python
df.groupby("customer_state")["total_payment"].sum().sort_values(ascending=False).plot(kind='bar', figsize=(8,5))

plt.xlabel("Customer_State")
plt.ylabel("total_payment")
plt.title("Total Payment per Customer State")

plt.gca().yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))

plt.show()
```

```python
df.groupby("category")["price"].sum().sort_values(ascending=False).plot(kind='bar', figsize=(8,5))

plt.xlabel("Category")
plt.ylabel("Price")
plt.title("Total Price per Category ")

plt.gca().yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))

plt.show()

```

```python
df.groupby("category")["total_payment"].sum().sort_values(ascending=False).plot(kind='bar', figsize=(8,5))

plt.xlabel("Category")
plt.ylabel("Total Payment")
plt.title("Total Payment per Category ")

plt.gca().yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))

plt.show()

```
