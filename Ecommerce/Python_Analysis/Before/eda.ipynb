```python
df['price_bin'] = pd.cut(
    df['price'],
    bins=[0,1000,3000,6000,10000],
    labels=['<1000','1000-3000','3000-6000','6000-10000']
)


df.groupby('payment_installments')['price_bin'] \
  .value_counts() \
  .plot(kind='hist', figsize=(8,5))

plt.title('Price Count (Binned) by Payment Installments')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.scatter(df['price'], df['review_score'], alpha=0.5)
plt.xlabel('Price')
plt.ylabel('Review Score')
plt.title('Price vs Review Score')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.scatter(df['payment_installments'], df['price'], alpha=0.5)
plt.xlabel('Payment Installments')
plt.ylabel('Price')
plt.title('Payment Installments vs Price')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
df.boxplot(column='review_score', by='payment_type')
plt.xlabel('Payment Type')
plt.ylabel('Review Score')
plt.title('Review Score by Payment Type')
plt.suptitle('')  # extra default title remove
plt.show()
```

```python

import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.scatter(df['freight_value'], df['review_score'], alpha=0.5)
plt.xlabel('Freight Value (Shipping Cost)')
plt.ylabel('Review Score')
plt.title('Freight Value vs Review Score')
plt.show()
```

```python
import pandas as pd

df['delivery_delay_days'] = (
    pd.to_datetime(df['order_delivered_customer_date']) -
    pd.to_datetime(df['order_estimated_delivery_date'])
).dt.days
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.scatter(df['delivery_delay_days'], df['review_score'], alpha=0.5)
plt.xlabel('Delivery Delay (Days)')
plt.ylabel('Review Score')
plt.title('Delivery Delay vs Review Score')
plt.axvline(0)   # on-time vs late divider
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.scatter(df['product_weight_g'], df['freight_value'], alpha=0.5)
plt.xlabel('Product Weight (grams)')
plt.ylabel('Freight Value (Shipping Cost)')
plt.title('Product Weight vs Freight Value')
plt.show()

```
